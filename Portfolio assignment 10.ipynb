```python
import seaborn as sns
import pandas as pd
```

```python
songs = pd.read_csv('songs.csv', index_col='Unnamed: 0', quotechar="`") 
```

```python
songs
```

```python
songsDataCorrelations = songs.corr()
songsDataCorrelations.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```

```python
songs.plot(kind='scatter', x='dB', y='Energy')
```

Tussen het aantal dB en Energy zit een correlatie van 0.54, dit is de sterkste positieve correlatie uit deze database. Zoals in het scatter plot getoond wordt het energielevel verhoogd naarmate het aantal dB dichterbij 0 komt. Dit betekent dat energieke nummers vaak luider zijn dan rustigere nummers. 

```python
songs.plot(kind='scatter', x='Acousticness', y='Energy')
```

Tussen de acousticness en de energie zit ook een correlatie. Dit is echter een negatieve correlate. Naarmate de acousticness verhoogd, daalt het energielevel van het nummer. Dit komt omdat akoustische nummers vaak langzamere nummers zijn.
