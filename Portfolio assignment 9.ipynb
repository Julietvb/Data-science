```python
import seaborn as sns
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
penguinDataCorrelations = penguins.corr()
penguinDataCorrelations.style.background_gradient(cmap='coolwarm', axis=None).format(precision=2)
```
